```python
from dotenv import load_dotenv
load_dotenv()  # take environment variables from .env.
```

```python
from anthropic import Anthropic
from typing import Final


MODEL: Final[str] = "claude-sonnet-4-6"
USER_ROLE: Final[str] = "user"
ASSISTANT_ROLE: Final[str] = "assistant"


client = Anthropic()

def _send_message_to_claude(messages: list[dict]) -> str:
    message = client.messages.create(
        model=MODEL,
        max_tokens=1000,
        messages=messages
    )
    return message

def chat_with_claude():
    messages = [{"role": USER_ROLE, "content": "Hi claude"}]
    while True:
        response = _send_message_to_claude(messages)
        print(f"Claude: {response.content[0].text}")
        messages.append({"role": ASSISTANT_ROLE, "content": response.content[0].text})
        try:
            user_input = input("You: ")
        except (EOFError, KeyboardInterrupt):
            print("\nExiting chat.")
            break
        if not user_input.strip():
            print("Exiting chat.")
            break
        messages.append({"role": USER_ROLE, "content": user_input})

chat_with_claude()
```

```python
"""Streaming mode example"""

messages = [{"role": USER_ROLE, "content": "Hi claude. Tell me 3 sentences about yourself."}]

with client.messages.stream(
    model=MODEL,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")
    print("++++Final stream response message++++")
    print(stream.get_final_message())  # Wait for the final response to be received
```
